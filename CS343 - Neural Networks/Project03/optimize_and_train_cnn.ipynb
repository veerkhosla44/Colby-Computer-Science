## Task 5: Implement weight optimizers for gradient descent

To change the weights during training, we need an optimization algorithm to have our loss decrease over epochs as we learn the structure of the input patterns. Until now, we used **Stochastic gradient descent (SGD)**, which is the simplest algorithm. You will implement 4 popular algorithms:

1. `SGD` (stochastic gradient descent)
2. `SGD_Momentum` (stochastic gradient descent with momentum)
3. `Adam` (Adaptive Moment Estimation)
4. `AdamW` (Adaptive Moment Estimation Weight decay) <br/>*Skip AdamW for now. There is a task below that focuses on this optimizer.*

Implement each of these according to the update equations (the `update_weights()` in each subclass in `optimizer.py`). Let's use $w_t$ in the math below to represent the weights in a layer at time step $t$, $dw$ to represent the gradient of the weights in a layer, and $\eta$ represent the learning rate. We use vectorized notation below (update applies to all weights element-wise). Then:

**SGD**: 

$w_{t} = w_{t-1} - \eta \times dw$

**SGD (momentum)**:

$v_{t} = m \times v_{t-1} - \eta \times dw$

$w_{t} = w_{t-1} + v_t$

where $v_t$ is called the `velocity` at time $t$. At the first time step (0), velocity should be set to all zeros and have the same shape as $w$. $m$ is a constant that determines how much of the gradient obtained on the previous time step should factor into the weight update for the current time step.


**Adam**:

$v_{t} = \beta_1 \times v_{t-1} + (1 - \beta_1)\times dw$

$p_{t} = \beta_2 \times p_{t-1} + (1 - \beta_2)\times dw^2$

$vc = v_{t} / \left (1-(\beta_1^t) \right )$

$pc = p_{t} / \left (1-(\beta_2^t) \right )$

$w_{t} = w_{t-1} - \left ( \eta \times vc \right ) / \left ( \sqrt(pc) + \epsilon \right ) $


Like SGD (momentum), Adam records momentum terms $v$ and $p$. At time step 0, you should initialize them to zeros in an array equal in size to the weights. $vc$ and $pc$ are variables computed on each time step. The remaining quantities are constants. Note that $t$ keeps track of the integer time step, and needs to be incremented on each update.

### 8e. Visualize layer weights

Run the following code and submit the inline image of the weight visualization of the 1st layer (convolutional layer) of the network.

**Note:**
- Setting optional parameter to `True` will let you save a .PNG file in your project folder of your weights. I'd suggest setting it to `False` unless look at your weights and they look like they are worth saving. You don't want a training run that produces undesirable weights to overwrite your good looking results!

```python
def plot_weights(wts, saveFig=True, filename='convWts.png'):
    grid_sz = int(np.sqrt(len(wts)))
    plt.figure(figsize=(10,10))
    for x in range(grid_sz):
        for y in range(grid_sz):
            lin_ind = np.ravel_multi_index((x, y), dims=(grid_sz, grid_sz))
            plt.subplot(grid_sz, grid_sz, lin_ind+1)
            currImg = wts[lin_ind]
            low, high = np.min(currImg), np.max(currImg)
            currImg = 255*(currImg - low) / (high - low)
            currImg = currImg.astype('uint8')
            plt.imshow(currImg)
            plt.gca().axis('off')
    plt.subplots_adjust(wspace=0.01, hspace=0.01)
    if saveFig:
        plt.savefig(filename)
    plt.show()
```

```python
dropout_rates = [0.0, 0.1, 0.5, 0.9]

loss_histories = {}
test_accuracies = {}

for rate in dropout_rates:
    print(f"Training with dropout rate: {rate}")
    
    net = ConvNet4AccelV2(
        input_shape=(3, 32, 32),
        n_kers=(32,),
        pooling_sizes=(2,),
        pooling_strides=(2,),
        n_classes=10,
        dropout_rate=rate,
        r_seed=0,
        verbose=False,
        dense_interior_units=(200,)
    )
    
    net.compile('adam')
    net.fit(
        x_train,
        y_train,
        x_val,
        y_val,
        acc_freq=10,
        print_every=10,
        n_epochs=5
    )
    
    loss_histories[rate] = net.loss_history
    test_accuracy = net.accuracy(x_test, y_test)
    test_accuracies[rate] = test_accuracy
    print(f"Test accuracy with dropout {rate}: {test_accuracy:.4f}")
```

```python
# Plot training loss for each dropout rate
plt.figure(figsize=(10, 6))
for rate, losses in loss_histories.items():
    plt.plot(losses, label=f"Dropout {rate}")
plt.title("Training Loss vs. Epoch for Different Dropout Rates")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# Plot test accuracy for each dropout rate
plt.figure(figsize=(8, 5))
plt.bar(range(len(dropout_rates)), test_accuracies.values(), tick_label=[f"{rate}" for rate in dropout_rates])
plt.title("Test Accuracy vs. Dropout Rate")
plt.xlabel("Dropout Rate")
plt.ylabel("Test Accuracy")
plt.grid(True)
plt.show()

print("Dropout Rates, Final Losses, and Test Accuracies")
print("-" * 50)
for rate in dropout_rates:
    print(f"Dropout Rate: {rate:.1f} | Final Loss: {loss_histories[rate][-1]:.4f} | Test Accuracy: {test_accuracies[rate]:.2f}%")
```

Below, train separate `ConvNet4AccelV2` (with dropout) nets with separate 4 regularization strengths `[0, 0.01, 0.1, 0.5]`. As in your dropout experiment, each net should use a different regularization value. In each case, train the net once with Adam and once again with AdamW. Compute/keep track of the test accuracy achieved by each net on STL-10.

```python
# Regularization strengths to test
regularization_strengths = [0, 0.01, 0.1, 0.5]

# Storage for results
adam_accuracies = []
adamw_accuracies = []

# Loop through regularization strengths for both optimizers
for reg in regularization_strengths:
    print(f"Training with Adam optimizer (reg={reg})")
    net_adam = ConvNet4AccelV2(input_shape=(3, 32, 32), r_seed=0, reg=reg, verbose=False, dense_interior_units=(200,), dropout_rate=0.2)
    net_adam.compile('adam')
    net_adam.fit(
        x_train,
        y_train,
        x_val,
        y_val,
        acc_freq=10,
        print_every=10,
        n_epochs=10
    )
    adam_accuracy = net_adam.accuracy(x_test, y_test)
    adam_accuracies.append(adam_accuracy)

    print(f"Training with AdamW optimizer (reg={reg})")
    net_adamw = ConvNet4AccelV2(input_shape=(3, 32, 32), r_seed=0, reg=reg, verbose=False, dense_interior_units=(200,), dropout_rate=0.2)
    net_adamw.compile('adamw', reg=reg)
    net_adamw.fit(
        x_train,
        y_train,
        x_val,
        y_val,
        acc_freq=10,
        print_every=10,
        n_epochs=10,
    )
    adamw_accuracy = net_adamw.accuracy(x_test, y_test)
    adamw_accuracies.append(adamw_accuracy)

# Print results for clarity
for i, reg in enumerate(regularization_strengths):
    print(f"Reg={reg}: Adam Test Accuracy={adam_accuracies[i]:.4f}, AdamW Test Accuracy={adamw_accuracies[i]:.4f}")
```

### Extension 1: Confusion Matrix

```python
# Define the class labels
label_names = [
    "airplane", "bird", "car", "cat", 
    "deer", "dog", "horse", "monkey", 
    "ship", "truck"
]

# Get predictions for the test set
predictions = net_adamw.predict(x_test)

def compute_confusion_matrix(true_labels, predicted_labels, num_classes):
    conf_matrix = np.zeros((num_classes, num_classes), dtype=int)
    for t, p in zip(true_labels, predicted_labels):
        conf_matrix[t, p] += 1
    return conf_matrix


num_classes = len(label_names)

# Compute the confusion matrix
conf_matrix = compute_confusion_matrix(y_test, predictions, num_classes)

# Print the confusion matrix
print("Confusion Matrix:\n", conf_matrix)

# Display the labels along with the confusion matrix
print("\nClass Labels:")
for idx, label in enumerate(label_names):
    print(f"{idx}: {label}")

# Plot the confusion matrix
plt.figure(figsize=(10, 8))
plt.imshow(conf_matrix, cmap='viridis')

plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

# Add colorbar and grid for better visualization
plt.colorbar()

# Set tick labels to class names
plt.xticks(ticks=np.arange(num_classes), labels=label_names, rotation=45, ha='right')
plt.yticks(ticks=np.arange(num_classes), labels=label_names)

# Annotate each cell with the count
for i in range(num_classes):
    for j in range(num_classes):
        plt.text(j, i, conf_matrix[i, j], ha='center', va='center', color='white')

plt.grid(False)
plt.tight_layout()
plt.show()
```
